категорий 4, 12 и 21 за период с 2005 по 2010 года. 
Не учитывайте в расчетах отрицательные значения quantity.

```python
#data = pd.read_csv('power.csv')
# array_ = data.country.unique()
# array_
#data.head()
```

```python
data = pd.read_csv('power.csv')
filtered_countries = data[(
    (data.quantity >= 0) &
    (data.year.isin(['2005','2006','2007','2008','2009','2010'])) &
    (data.category.isin(['4','12','21'])) &
    (data.country.isin(['Estonia','Lithuania','Latvia']))
 )]
filtered_countries.quantity.sum()
```
